```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. ÿÆŸàÿßŸÜÿØŸÜ ÿØÿßÿØŸá
df = pd.read_csv("student_depression_dataset.csv")

# 2. ÿ¨ÿØÿß ⁄©ÿ±ÿØŸÜ ÿ®ÿ±⁄Üÿ≥ÿ® ŸáÿØŸÅ Ÿà Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß
X = df.drop("Depression", axis=1)  # ÿ≠ÿ∞ŸÅ ÿ≥ÿ™ŸàŸÜ Depression ÿßÿ≤ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß
y = df["Depression"]  # ÿ®ÿ±⁄Üÿ≥ÿ® ŸáÿØŸÅ

# 3. ÿ™ÿ®ÿØ€åŸÑ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å ÿ∫€åÿ±ÿπÿØÿØ€å ÿ®Ÿá ÿπÿØÿØ€å (One-hot encoding)
X = pd.get_dummies(X)  # ÿ™ÿ®ÿØ€åŸÑ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å ÿ∫€åÿ±ÿπÿØÿØ€å ÿ®Ÿá Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å ÿπÿØÿØ€å

# 4. ÿ™ŸÇÿ≥€åŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß ÿ®Ÿá ÿ¢ŸÖŸàÿ≤ÿ¥ Ÿà ÿ™ÿ≥ÿ™ (75/25)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# 5. ÿ™ÿπÿ±€åŸÅ ŸÖÿØŸÑ Ÿæÿß€åŸá Random Forest
rf_model = RandomForestClassifier(random_state=42)

# 6. ÿ™ÿπÿ±€åŸÅ Ÿæÿßÿ±ÿßŸÖÿ™ÿ±Ÿáÿß€å€å ⁄©Ÿá ŸÖ€å‚ÄåÿÆŸàÿß€åŸÖ ÿ™ÿ≥ÿ™ ⁄©ŸÜ€åŸÖ
param_grid = {
    'n_estimators': [50, 100, 200],  # ÿ™ÿπÿØÿßÿØ ÿØÿ±ÿÆÿ™‚ÄåŸáÿß
    'max_features': ['sqrt', 'log2', None]  # ÿ™ÿπÿØÿßÿØ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å Ÿáÿ± ÿØÿ±ÿÆÿ™
}

# 7. ÿ≥ÿßÿÆÿ™ GridSearchCV ÿ®ÿ±ÿß€å ÿ¨ÿ≥ÿ™ÿ¨Ÿà€å ÿ®Ÿáÿ™ÿ±€åŸÜ ÿ™ÿ±⁄©€åÿ® Ÿæÿßÿ±ÿßŸÖÿ™ÿ±
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5,  # ÿ™ÿπÿØÿßÿØ ÿ®ÿÆÿ¥‚ÄåŸáÿß€å cross-validation
    scoring='accuracy',
    n_jobs=-1  # ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ŸáŸÖŸá Ÿáÿ≥ÿ™Ÿá‚ÄåŸáÿß€å CPU ÿ®ÿ±ÿß€å ÿ≥ÿ±ÿπÿ™ ÿ®€åÿ¥ÿ™ÿ±
)

# 8. ÿßÿ¨ÿ±ÿß€å ÿ¨ÿ≥ÿ™ÿ¨Ÿà ÿ±Ÿà€å ÿØÿßÿØŸá ÿ¢ŸÖŸàÿ≤ÿ¥
grid_search.fit(X_train, y_train)

# 9. ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿ®Ÿáÿ™ÿ±€åŸÜ ŸÖÿØŸÑ Ÿà ŸÜÿ™ÿß€åÿ¨
best_model = grid_search.best_estimator_  # ÿ®Ÿáÿ™ÿ±€åŸÜ ŸÖÿØŸÑ Ÿæ€åÿØÿß ÿ¥ÿØŸá
best_params = grid_search.best_params_  # ÿ®Ÿáÿ™ÿ±€åŸÜ Ÿæÿßÿ±ÿßŸÖÿ™ÿ±Ÿáÿß
best_cv_score = grid_search.best_score_  # ÿØŸÇÿ™ ÿ®Ÿáÿ™ÿ±€åŸÜ ŸÖÿØŸÑ ÿØÿ± ⁄©ÿ±ÿßÿ≥‚ÄåŸàŸÑ€åÿØ€åÿ¥ŸÜ

print("\033[93müèÜ ÿ®Ÿáÿ™ÿ±€åŸÜ Ÿæÿßÿ±ÿßŸÖÿ™ÿ±Ÿáÿß:\033[0m", best_params)
print(f"\033[92m‚úÖ ÿØŸÇÿ™ ŸÖ€åÿßŸÜ⁄Ø€åŸÜ ⁄©ÿ±ÿßÿ≥‚ÄåŸàŸÑ€åÿØ€åÿ¥ŸÜ: {best_cv_score:.2%}\033[0m")

# 10. ÿßÿ±ÿ≤€åÿßÿ®€å ŸÜŸáÿß€å€å ŸÖÿØŸÑ ÿ±Ÿà€å ÿØÿßÿØŸá ÿ™ÿ≥ÿ™
test_accuracy = best_model.score(X_test, y_test)  # ÿßÿ±ÿ≤€åÿßÿ®€å ÿØŸÇÿ™ ŸÖÿØŸÑ ÿ±Ÿà€å ÿØÿßÿØŸá ÿ™ÿ≥ÿ™
print(f"\033[96müéØ ÿØŸÇÿ™ ŸÜŸáÿß€å€å ÿ±Ÿà€å ÿØÿßÿØŸá ÿ™ÿ≥ÿ™: {test_accuracy:.2%}\033[0m")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# ŸÖÿ±ÿ≠ŸÑŸá: Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿØŸàÿ®ÿßÿ±Ÿá ÿ®ÿß ÿ®Ÿáÿ™ÿ±€åŸÜ ŸÖÿØŸÑ (ŸÅŸÇÿ∑ ÿ®ÿ±ÿß€å ÿßÿ∑ŸÖ€åŸÜÿßŸÜ)
# ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ŸÖÿØŸÑ ÿ®Ÿáÿ™ÿ±€åŸÜ (best_model) ÿ®ÿ±ÿß€å Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ™ÿ≥ÿ™
y_pred = best_model.predict(X_test)

# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ŸÖÿßÿ™ÿ±€åÿ≥ ÿ≥ÿ±ÿØÿ±⁄ØŸÖ€å (confusion matrix) ÿ®ÿß ŸÖŸÇÿß€åÿ≥Ÿá Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å‚ÄåŸáÿß Ÿà ÿ®ÿ±⁄Üÿ≥ÿ®‚ÄåŸáÿß€å ŸàÿßŸÇÿπ€å
cm = confusion_matrix(y_test, y_pred)

# ÿ±ÿ≥ŸÖ ⁄Øÿ±ÿßŸÅ€å⁄©€å ŸÖÿßÿ™ÿ±€åÿ≥ ÿ≥ÿ±ÿØÿ±⁄ØŸÖ€å
# ŸÜŸÖÿß€åÿ¥ ŸÖÿßÿ™ÿ±€åÿ≥ ÿ≥ÿ±ÿØÿ±⁄ØŸÖ€å ÿ®ÿß ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ŸÜŸÖÿß€åÿ¥⁄Øÿ± ŸÖŸàÿ¨ŸàÿØ ÿØÿ± sklearn
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Depression", "Depression"])

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ± ŸÖÿßÿ™ÿ±€åÿ≥ ÿ≥ÿ±ÿØÿ±⁄ØŸÖ€å ÿ®ÿß ÿ±ŸÜ⁄Ø‚Äåÿ®ŸÜÿØ€å ÿ¢ÿ®€å Ÿà ŸÅÿ±ŸÖÿ™ ÿπÿØÿØ€å ÿØÿßÿÆŸÑ Ÿáÿ± ÿÆÿßŸÜŸá
disp.plot(cmap='Blues', values_format='d')

# ÿ™ŸÜÿ∏€åŸÖ ÿπŸÜŸàÿßŸÜ ŸÜŸÖŸàÿØÿßÿ±
plt.title("Confusion Matrix - Depression Prediction")

# ŸÜŸÖÿß€åÿ¥ ŸÜŸÖŸàÿØÿßÿ±
plt.show()
```
