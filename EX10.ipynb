```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. خواندن داده
df = pd.read_csv("student_depression_dataset.csv")

# 2. جدا کردن برچسب هدف و ویژگی‌ها
X = df.drop("Depression", axis=1)  # حذف ستون Depression از ویژگی‌ها
y = df["Depression"]  # برچسب هدف

# 3. تبدیل ویژگی‌های غیرعددی به عددی (One-hot encoding)
X = pd.get_dummies(X)  # تبدیل ویژگی‌های غیرعددی به ویژگی‌های عددی

# 4. تقسیم داده‌ها به آموزش و تست (75/25)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# 5. تعریف مدل پایه Random Forest
rf_model = RandomForestClassifier(random_state=42)

# 6. تعریف پارامترهایی که می‌خوایم تست کنیم
param_grid = {
    'n_estimators': [50, 100, 200],  # تعداد درخت‌ها
    'max_features': ['sqrt', 'log2', None]  # تعداد ویژگی‌های هر درخت
}

# 7. ساخت GridSearchCV برای جستجوی بهترین ترکیب پارامتر
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5,  # تعداد بخش‌های cross-validation
    scoring='accuracy',
    n_jobs=-1  # استفاده از همه هسته‌های CPU برای سرعت بیشتر
)

# 8. اجرای جستجو روی داده آموزش
grid_search.fit(X_train, y_train)

# 9. استخراج بهترین مدل و نتایج
best_model = grid_search.best_estimator_  # بهترین مدل پیدا شده
best_params = grid_search.best_params_  # بهترین پارامترها
best_cv_score = grid_search.best_score_  # دقت بهترین مدل در کراس‌ولیدیشن

print("\033[93m🏆 بهترین پارامترها:\033[0m", best_params)
print(f"\033[92m✅ دقت میانگین کراس‌ولیدیشن: {best_cv_score:.2%}\033[0m")

# 10. ارزیابی نهایی مدل روی داده تست
test_accuracy = best_model.score(X_test, y_test)  # ارزیابی دقت مدل روی داده تست
print(f"\033[96m🎯 دقت نهایی روی داده تست: {test_accuracy:.2%}\033[0m")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# مرحله: پیش‌بینی دوباره با بهترین مدل (فقط برای اطمینان)
# استفاده از مدل بهترین (best_model) برای پیش‌بینی داده‌های تست
y_pred = best_model.predict(X_test)

# محاسبه ماتریس سردرگمی (confusion matrix) با مقایسه پیش‌بینی‌ها و برچسب‌های واقعی
cm = confusion_matrix(y_test, y_pred)

# رسم گرافیکی ماتریس سردرگمی
# نمایش ماتریس سردرگمی با استفاده از نمایشگر موجود در sklearn
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Depression", "Depression"])

# رسم نمودار ماتریس سردرگمی با رنگ‌بندی آبی و فرمت عددی داخل هر خانه
disp.plot(cmap='Blues', values_format='d')

# تنظیم عنوان نمودار
plt.title("Confusion Matrix - Depression Prediction")

# نمایش نمودار
plt.show()
```
