```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# بارگذاری داده‌ها از فایل CSV و حذف ستون "id"
df = pd.read_csv("student_depression_dataset.csv").drop(columns=["id"])

# تنظیم ظاهر Seaborn برای ایجاد شبکه سفید
sns.set(style="whitegrid")

# رسم نمودار scatter با استفاده از seaborn
plt.figure(figsize=(8, 6))  # تنظیم اندازه نمودار
sns.scatterplot(
    data=df,  # داده‌ها
    x="Study Satisfaction",  # ویژگی محور X
    y="Job Satisfaction",  # ویژگی محور Y
    hue="Depression",  # گروه‌بندی بر اساس برچسب "Depression"
    palette={0: "green", 1: "red"},  # رنگ‌ها: سبز برای افرادی که افسردگی ندارند و قرمز برای افرادی که افسردگی دارند
    alpha=0.6  # شفافیت نقاط نمودار
)

# تنظیمات ظاهری نمودار
plt.title("Depression Based on Study vs Job Satisfaction")  # عنوان نمودار
plt.legend(title="Depression")  # عنوان برای برچسب‌ها در نمودار
plt.tight_layout()  # تنظیمات برای جلوگیری از تداخل برچسب‌ها
plt.show()  # نمایش نمودار
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# بارگذاری داده‌ها و حذف ستون "id"
df = pd.read_csv("student_depression_dataset.csv").drop(columns=["id"])

# فیلتر کردن داده‌ها برای افرادی که افسردگی دارند
depressed = df[df["Depression"] == 1]

# تنظیم اندازه نمودار
plt.figure(figsize=(8, 6))

# رسم نمودار KDE (توزیع هسته‌ای) برای Study Satisfaction و Job Satisfaction
sns.kdeplot(
    data=depressed,               # داده‌های فیلتر شده برای افسردگی
    x="Study Satisfaction",       # ویژگی محور X
    y="Job Satisfaction",         # ویژگی محور Y
    cmap="Reds",                  # استفاده از رنگ‌های قرمز برای نقاط با تراکم بالا
    fill=True,                    # پر کردن نواحی با رنگ
    thresh=0.05,                  # آستانه‌ای برای تعیین نقاط با تراکم کم که نمایش داده نشوند
    levels=10                     # تعداد سطوح رنگی برای نمایش تفاوت‌ها
)

# تنظیمات ظاهری نمودار
plt.title("Study vs Job Satisfaction")  # عنوان نمودار
plt.tight_layout()  # تنظیمات برای جلوگیری از تداخل برچسب‌ها
plt.show()  # نمایش نمودار
```
